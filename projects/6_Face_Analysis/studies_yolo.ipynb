```python
from ultralytics import YOLO
import cv2

# carrega modelo pronto (já treinado)
model = YOLO("yolov8n.pt")

# lê imagem
img = cv2.imread("images/px-girl.jpg")

# faz detecção
results = model(img)

# mostra resultado
annotated = results[0].plot()
cv2.imshow("YOLO", annotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
